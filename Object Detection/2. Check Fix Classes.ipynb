```python
from collections import Counter
import pandas as pd
import os
```

```python
ROOT_PATH = '/content/drive/MyDrive/Omdena/Challenge: Disease Detection in Coffee Plants/Object Detection/Dataset/labels'
```

```python
filename = []
results = []
```

```python
for text in sorted(os.listdir(ROOT_PATH)):
  f = open(os.path.join(ROOT_PATH, text), 'r')
  file_contents = f.readlines()

  for row in file_contents:
    filename.append(text)
    results.append(row)
```

```python
class_for_each_annotation = [x.split(' ')[0] for x in results]
```
